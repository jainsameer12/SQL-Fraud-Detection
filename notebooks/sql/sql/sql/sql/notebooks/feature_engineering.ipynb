```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../data/fraud.db")
```

```python
burst = pd.read_sql_query(open("../sql/burst_pattern.sql").read(), conn)
amount = pd.read_sql_query(open("../sql/amount_deviation.sql").read(), conn)
device = pd.read_sql_query(open("../sql/device_switch.sql").read(), conn)
location = pd.read_sql_query(open("../sql/location_switch.sql").read(), conn)
```

```python
features = burst.merge(amount, on="user_id") \
                .merge(device, on="user_id") \
                .merge(location, on="user_id")
```

```python
labels = pd.read_sql_query("""
SELECT user_id, MAX(is_fraud) AS is_fraud
FROM transactions
GROUP BY user_id
""", conn)

features = features.merge(labels, on="user_id")
features.head()
```

```python
features.to_csv("../data/processed/fraud_features.csv", index=False)
conn.close()

print("Feature table saved successfully")
```
